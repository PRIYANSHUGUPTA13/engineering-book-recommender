## Popularity based


```python
ratings.merge(Engineering_books,on='ISBN')
```

```python
ratings.merge(Engineering_books,on='ISBN').shape
```

```python
ratings_with_name=ratings.merge(Engineering_books,on='ISBN')
```

```python
num_rating_df= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df= ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df[popular_df['num_ratings']>=200].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df=popular_df.merge(Engineering_books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_ratings','avg_ratings']]
```

## Collabrative filtering based


```python
X=ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 1
padhe_likhe_users = X[X].index
```

```python
ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
filtered_rating=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=35
famous_books = y[y].index
```

```python
famous_books
```

```python
 final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
 pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores[0]
```

```python
def recommend(book_name, books_df):
    try:
        index = np.where(pt.index == book_name)[0][0]
    except IndexError:
        print(f"Sorry, the book '{book_name}' was not found.")
        return []

    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]
   
    data = []
    
    for i in similar_items:
        item = []
        temp_df = books_df[books_df['Book-Title'] == pt.index[i[0]]]
        item.append(temp_df.drop_duplicates('Book-Title')['Book-Title'].values[0])
        item.append(temp_df.drop_duplicates('Book-Title')['Book-Author'].values[0])
        item.append(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values[0])
        data.append(item)
        
    return data
```

```python
recommend('Control Systems Engineering', Engineering_books)
```
